```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.geeksforgeeks.org/python-programming-language/'
res = req.get(url)   # type of response: <class 'requests.models.Response'>
print(res.content)
```

```python
soup = bs(res.content,'html.parser')
print(soup.prettify())
```

```python
divs = soup.div
divs_a = soup.div.a
divs
```

```python
class_x = soup.find('div',class_='button')
class_x
```

```python
strs = soup.find(string='Mark all as read')
strs
```
