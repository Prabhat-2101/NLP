# Type-Token Ratio: it is the ratio of unique words by total number of words in a corpus.

1. Finding Type-Token Ratio without any preprocessing.

```python
with open('demo_corpus.txt','r') as fp:
    data = fp.read()
    words = list(data.split())
    unique_words = []
    for word in words:
        if word not in unique_words:
            unique_words.append(word)
    token = len(words)
    types = len(unique_words)
    ttr = types/token*1.0
    print("Type-Token Ratio: ",ttr)
```

2. Finding Type-Token Ratio after removing special chars and lowercasing.

```python
import re
with open('demo_corpus.txt','r') as fp:
    data = fp.read()
    data = re.sub(r'[,(){}]','',data)
    words = list(data.split())
    lower_words = [i.lower() for i in words]
    unique_words = []
    for word in lower_words:
        if word not in unique_words:
            unique_words.append(word)
    token = len(lower_words)
    types = len(unique_words)
    ttr = types/token*1.0
    print("Type-Token Ratio: ",ttr)
```
